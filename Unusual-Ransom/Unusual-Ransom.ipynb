We can manually see the outliers and create a range by listing them in a sorted order.
This only works easily due to the small sample size
the range was determined as:
x 280 to 758
y -300 to -119

```python
# Find each row that is an outlier and add to a list to delete
to_remove = []
for index, row in df.iterrows():
    if row['x'] < 280 or row['x'] > 758:
        to_remove.append(index)
    elif row['y'] < -300 or row['y'] > -119:
        to_remove.append(index)
print("We are to remove " + str(len(to_remove)) + " outliers")
```

```python
# Create the new data set from the exisiting
new_data = df.drop(to_remove, axis=0)
```

```python
plt.scatter(new_data.x, new_data.y);
```
